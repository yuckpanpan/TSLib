```python
import os
import numpy as np

data_dir = '/home/wangpenglei/PowerPred/data/'
suffix = '14_14_with_Y_PV'
def get_data_or_fail(data_dir: str, suffix='', real=False):
    data_path = os.path.join(data_dir, f"wea_data_{suffix}.npz")

    if not (os.path.exists(data_path)):
        raise FileNotFoundError(f"Data files not found in {data_dir}: {data_path}")
    data = np.load(data_path, allow_pickle=True)
    X = data['X']
    if real:
        Y = data['Y_real']
    else:
        Y = data['Y_before']

    return X, Y, data['t0_idx'], data['t0s']
X, Y, t0_idx, t0s = get_data_or_fail(data_dir, suffix) 
```
